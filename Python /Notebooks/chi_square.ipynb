```python
# Install ISLR package to get the Default dataset
!pip install ISLP

# Import necessary libraries
import pandas as pd
import numpy as np
from ISLP import load_data
from scipy.stats import chi2_contingency

# Load the Default dataset
df = load_data('Default')

# Display the first few rows of the dataset
print(df.head())

# Create a contingency table for default (Yes/No) vs student (Yes/No)
table_data = pd.crosstab(df['default'], df['student'])
print("\nContingency Table:")
print(table_data)

# Perform the Chi-Square Test
chi2, p, dof, expected = chi2_contingency(table_data)

# Display the test results
print(f"\nChi-Square Statistic: {chi2}")
print(f"p-value: {p}")
print(f"Degrees of Freedom: {dof}")
print("\nExpected Frequencies:")
print(pd.DataFrame(expected, index=table_data.index, columns=table_data.columns))
```
