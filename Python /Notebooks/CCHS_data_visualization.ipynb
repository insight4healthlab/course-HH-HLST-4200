Read Tab To Dataframe

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Define the data path (update this if the file is in a different location)
data_path = 'course-HH-HLST-4200/datasets/cchs-82M0013-E-2017-2018-Annual-component_F1-subset.tab'

data_path = '/content/gdrive/MyDrive/' + data_path
# Read the .tab file into a pandas DataFrame
def read_tab_file(data_path):
    try:
        # Using pandas to read the tab-separated file
        df = pd.read_csv(data_path, sep='\t', engine='python')
        print("File successfully loaded into a DataFrame.")
        return df
    except FileNotFoundError:
        print(f"File not found at the path: {data_path}. Please verify the file location.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Call the function to load the file
data_frame = read_tab_file(data_path)

# Display the first few rows of the DataFrame
if data_frame is not None:
    print(data_frame.head())
```

```python
# Visualization: Basic Statistics
print("\nBasic statistics:")
print(data_frame.describe())
```

```python
# Visualization: Bar Plot for a Categorical Variable
plt.figure(figsize=(10, 6))
sns.countplot(y='DOMAM', data=data_frame, order=data_frame['DOMAM'].value_counts().index)
plt.title("Distribution of DOMAM")
plt.show()
```

```python
# Visualization: Histogram for a Numerical Variable
plt.figure(figsize=(10, 6))
sns.histplot(data_frame['MAM_005'], bins=20, kde=True)
plt.title("Histogram of MAM_005")
plt.xlabel("MAM_005")
plt.ylabel("Frequency")
plt.show()
```

```python
# Visualization: Scatter Plot for Numerical Variables
plt.figure(figsize=(10, 6))
sns.scatterplot(x='MAM_005', y='MAM_010', data=data_frame)
plt.title("Scatter Plot: MAM_005 vs MAM_010")
plt.xlabel("MAM_005")
plt.ylabel("MAM_010")
plt.show()
```

```python
# Visualization: Pairplot
plt.figure(figsize=(12, 8))
sns.pairplot(data_frame[['MAM_005', 'MAM_010', 'MAM_013A', 'MAM_013B']])
plt.title("Pairplot of Selected Numerical Columns")
plt.show()
```

```python
# Visualization: Correlation Heatmap
plt.figure(figsize=(12, 8))
corr = data_frame.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

```
