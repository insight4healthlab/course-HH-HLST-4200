```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Create the data frame
data = pd.DataFrame({
    'Years_Exp': [1.1, 1.3, 1.5, 2.0, 2.2, 2.9, 3.0, 3.2, 3.2, 3.7],
    'Salary': [39343.00, 46205.00, 37731.00, 43525.00, 39891.00,
               56642.00, 60150.00, 54445.00, 64445.00, 57189.00]
})

# Create the scatter plot
plt.scatter(data['Years_Exp'], data['Salary'], color='red')
plt.xlabel("Years Experienced")
plt.ylabel("Salary")
plt.title("Scatter Plot of Years Experienced vs Salary")
plt.show()

# Splitting the dataset into Training set and Test set
trainingset, testset = train_test_split(data, test_size=0.3, random_state=42)

# Fitting Simple Linear Regression to the Training set
regressor = LinearRegression()
regressor.fit(trainingset[['Years_Exp']], trainingset['Salary'])

# Summary of the model
print(f'Intercept: {regressor.intercept_}')
print(f'Coefficient: {regressor.coef_[0]}')

# Create a data frame with new input values
new_data = pd.DataFrame({'Years_Exp': [4.0, 4.5, 5.0]})

# Predict using the linear regression model
predicted_salaries = regressor.predict(new_data)

# Display the predicted salaries
print("Predicted Salaries:", predicted_salaries)

# Visualizing the Training set results
plt.scatter(trainingset['Years_Exp'], trainingset['Salary'], color='red', label='Training Data')
plt.plot(trainingset['Years_Exp'], regressor.predict(trainingset[['Years_Exp']]), color='blue', label='Regression Line')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of experience')
plt.ylabel('Salary')
plt.legend()
plt.show()

# Visualizing the Test set results
plt.scatter(testset['Years_Exp'], testset['Salary'], color='red', label='Test Data')
plt.plot(trainingset['Years_Exp'], regressor.predict(trainingset[['Years_Exp']]), color='blue', label='Regression Line')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
